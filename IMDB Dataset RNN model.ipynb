```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
max_features=10000
max_len=200
(x_train,y_train),(x_test,y_test)=imdb.load_data(num_words=max_features,maxlen=max_len)
```

```python
x_test, x_val, y_test, y_val = train_test_split(x_test, y_test, test_size=0.5, random_state=42)
```

```python
word_indexes=imdb.get_word_index()
len(word_indexes)
```

```python
reverse_word_index=dict([(value,key) for (key,value) in word_indexes.items()])
```

```python
for i in x_train[0]:
    if i != 0:  # Agar integer 0 nahi hai, tabhi print karein
        print(reverse_word_index.get(i + 3, "?"), end=' ')
```

```python
x_train=pad_sequences(x_train,maxlen=max_len)
x_test=pad_sequences(x_test,maxlen=max_len)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Embedding,Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
rnn_model = Sequential()
rnn_model.add(Embedding(input_dim=max_features, output_dim=32, input_length=max_len))
rnn_model.add(SimpleRNN(32))
rnn_model.add(Dropout(0.1))
rnn_model.add(Dense(1, activation='sigmoid'))

# EarlyStopping callback object
early_stopping_callback = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
rnn_model.fit(
    x_train,
    y_train,
    epochs=100,
    batch_size=64,
    validation_data=(x_val, y_val), # Yahan validation data use karein
    callbacks=[early_stopping_callback]
)
```
